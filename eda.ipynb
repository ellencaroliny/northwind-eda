```python
import matplotlib.pyplot as plt

# Converter a coluna de datas para datetime
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'], errors='coerce')

# Agrupar vendas por mês
df_orders['Mes'] = df_orders['order_date'].dt.month
vendas_por_mes = df_orders.groupby('Mes').size()

# Plotar gráfico de sazonalidade com eixos invertidos
plt.figure(figsize=(10, 6))
vendas_por_mes.plot(kind='barh', color='skyblue', alpha=0.8)
plt.title('Vendas por Mês', fontsize=16)
plt.ylabel('Mês', fontsize=12)
plt.xlabel('Quantidade de Vendas', fontsize=12)
plt.yticks(range(12), ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
